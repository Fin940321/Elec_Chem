```python
import numpy as np
import matplotlib.pyplot as plt
from cvsim.mechanisms import E_rev, E_q, E_qC, EE, SquareScheme
from cvsim.fit_curve import FitE_rev, FitE_q, FitE_qC, FitEE, FitSquareScheme
```

```python
# Example: The Square Scheme mechanism
potential, current = SquareScheme(
    start_potential=0.,        # V vs ref.
    switch_potential=-1.5,      # V vs ref.
    reduction_potential=-1.2,    # V vs ref.
    reduction_potential2=-0.7,  # V vs ref.
    scan_rate=1.0,              # V/s
    c_bulk=1,                   # mM
    diffusion_reactant=1e-5,    # cm2/s
    diffusion_product=1e-5,     # cm2/s
    alpha=0.5,                  # unitless
    alpha2=0.5,                 # unitless
    k0=5e-3,                    # cm/s
    k0_2=5e-3,                  # cm/s
    k_forward=0.3,              # 1/s
    k_backward=0.1,             # 1/s
    k_forward2=0.05,            # 1/s
    k_backward2=0.2,          # 1/s
).simulate()

fig,ax = plt.subplots()
ax.plot(potential, current*1000)
ax.set_xlabel('Potential (V vs ref.)')
ax.set_ylabel('Current (mA)')
```

```python
# increasing forward rate constant, kf_1
kf_1 = [1e-3, 1e-2, 1e-1, 1, 10]
potential = [[] for i in range(len(kf_1))]
current = [[] for i in range(len(kf_1))]
for i in range(len(kf_1)):
    potential[i], current[i] = SquareScheme(
        start_potential=0.0,        # V vs ref.
        switch_potential=-1.5,      # V vs ref.
        reduction_potential=-1.2,    # V vs ref.
        reduction_potential2=-0.7,  # V vs ref.
        scan_rate=1.0,              # V/s
        c_bulk=1,                   # mM
        diffusion_reactant=1e-5,    # cm2/s
        diffusion_product=1e-5,     # cm2/s
        alpha=0.5,                  # unitless
        alpha2=0.5,                 # unitless
        k0=5e-3,                    # cm/s
        k0_2=5e-3,                  # cm/s
        k_forward=float(kf_1[i]),              # 1/s
        k_backward=1e-2,             # 1/s
        k_forward2=0.05,            # 1/s
        k_backward2=0.2,          # 1/s
    ).simulate()

fig,ax = plt.subplots()
for i in range(len(kf_1)):
    ax.plot(potential[i], current[i]*1000, label="k$_{f1}$ = "+str(kf_1[i])+" cm/s")
#ax.plot(potential, current*1000)
ax.set_xlabel('Potential (V vs ref.)')
ax.set_ylabel('Current (mA)')
ax.legend()
```
